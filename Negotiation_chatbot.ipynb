```
import requests
from textblob import TextBlob


API_KEY = 'AIzaSyA6_QJNI2Kx9vyAoaFB00PGpul19U0hQ6A'
API_URL = 'https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash-latest:generateContent'

def get_input(prompt):
    return input(prompt)

def analyze_emotion(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0.1:
        return "positive"
    elif sentiment < -0.1:
        return "negative"
    else:
        return "neutral"

def get_ai_suggestion(prompt):
    headers = {
        'Content-Type': 'application/json',
    }
    data = {
        'contents': [
            {
                'parts': [
                    {'text': prompt}
                ]
            }
        ]
    }

    response = requests.post(f"{API_URL}?key={API_KEY}", headers=headers, json=data)

    if response.status_code == 200:
        return response.json()['candidates'][0]['content']['parts'][0]['text']
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return "Sorry, I couldn't get a response from the AI."

def negotiate_price(original_price, current_discount):
    while True:
        user_input = get_input(f"The actual price is ${original_price}. What is your offer? (e.g., 'I can offer $90') ")

        user_offer = None
        if '$' in user_input:
            try:
                user_offer = float(user_input.split('$')[1].strip())
            except ValueError:
                print("Invalid input. Please enter a valid offer with a dollar sign.")
                continue
        else:
            print("Please include a dollar amount in your offer (e.g., 'I can offer $90').")
            continue

        supplier_offer = max(original_price * (1 - current_discount / 100), user_offer + 2)
        print(f"Supplier's counter-offer: ${supplier_offer:.2f} (after {current_discount}% discount)")

        acceptance = get_input("Do you accept this offer? (yes/no) ")
        emotion = analyze_emotion(acceptance)
        print(f"User's sentiment: {emotion}")

        if acceptance.lower() == 'yes':
            print("Done deal!")
            break
        else:
            if emotion == "negative":
                current_discount += 3
            else:
                current_discount += 2
            print(f"Discount increased to {current_discount}%. Let's continue negotiating.")

            ai_prompt = f"User said: '{user_input}'. Suggest concise negotiation strategies or instructions."
            ai_response = get_ai_suggestion(ai_prompt)
            print(f"AI's suggestion: {ai_response}")


negotiate_price(100, 10)
```
